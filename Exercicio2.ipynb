````python
import random
N = 32
numeros = []
for i in range(N):
    numeros.append(random.randint(1, 1000))

#print(numeros)
````

````python

def busca_menor(numeros, inicio, j, menor, iteraco):
    if j == len(numeros):
        return menor, iteraco
    
    iteraco += 1
    if numeros[j] < numeros[menor]:
        menor = j
    
    return busca_menor(numeros, inicio, j + 1, menor, iteraco)


def ordena_semdiv_rec(numeros, i=0, iteraco=0):
    if i == len(numeros):
        print(iteraco)
        return iteraco
    
    menor, iteraco = busca_menor(numeros, i, i + 1, i, iteraco)
    
    aux = numeros[i]
    numeros[i] = numeros[menor]
    numeros[menor] = aux
    
    return ordena_semdiv_rec(numeros, i + 1, iteraco)

### 
# 32 numeros --> 0.0 segundos 496 iteracoes
# 2048 numeros --> 0.0 s 2096128 iteracoes 
# 1.048.576 numeros --> maximum recursion depth exceeded 

ordena_semdiv_rec(numeros, 0, 0)
numeros
````

   ```java
   long maxVal2(long A[], int init, int end) {

           return max(A[init], A[end]);

             int m = (init + end)/2;
   ```

````python
import random

def gera_vetor(nro_pares, nro_impares):
    res = None
    cont_par = 0
    cont_impar = 0

    if ((nro_pares >= 0) or
        (nro_impares >= 0) and
        (nro_pares + nro_impares > 0)):

        res = [0] * (nro_pares + nro_impares)

        while (cont_par < nro_pares) or (cont_impar < nro_impares):
            novo_num = random.randint(1, 98)

            if (novo_num % 2 == 0) and (cont_par < nro_pares):
                res[cont_par + cont_impar] = novo_num
                cont_par += 1

            elif (novo_num % 2 == 1) and (cont_impar < nro_impares):
                res[cont_par + cont_impar] = novo_num
                cont_impar += 1

    return res
````
