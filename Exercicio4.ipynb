```python
import time
import random

def multiply(x, y, n, stats):
    stats["iteracoes"] += 1

    if n == 1:
        return x * y
    else:
        m = (n + 1) // 2 

        a = x // (2 ** m)
        b = x % (2 ** m)
        c = y // (2 ** m)
        d = y % (2 ** m)

        e = multiply(a, c, m, stats)
        f = multiply(b, d, m, stats)
        g = multiply(b, c, m, stats)
        h = multiply(a, d, m, stats)

        return (2 ** (2 * m)) * e + (2 ** m) * (g + h) + f


def executar_teste(x, y, n):
    stats = {"iteracoes": 0}

    inicio = time.perf_counter()
    resultado = multiply(x, y, n, stats)
    fim = time.perf_counter()

    print(f"Teste com {n} bits")
    print(f"x = {x}")
    print(f"y = {y}")
    print(f"resultado algoritmo = {resultado}")
    print(f"resultado correto   = {x * y}")
    print(f"iteracoes = {stats['iteracoes']}")
    print(f"tempo = {fim - inicio:.10f} segundos")
    print("-" * 50)

#4 bits
x4 = random.randint(0, 2**4 - 1)
y4 = random.randint(0, 2**4 - 1)
executar_teste(x4, y4, 4)

#16 bits
x16 = random.randint(0, 2**16 - 1)
y16 = random.randint(0, 2**16 - 1)
executar_teste(x16, y16, 16)

#64 bits
x64 = random.randint(0, 2**64 - 1)
y64 = random.randint(0, 2**64 - 1)
executar_teste(x64, y64, 64)
```
