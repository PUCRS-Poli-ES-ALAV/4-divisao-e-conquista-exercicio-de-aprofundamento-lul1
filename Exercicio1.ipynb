```python
import random
import time

# contador global de iterações
iteracoes = 0

def merge(A, B):
    global iteracoes
    resultado = []
    i = 0
    j = 0

    while i < len(A) and j < len(B):
        iteracoes += 1
        if A[i] <= B[j]:
            resultado.append(A[i])
            i += 1
        else:
            resultado.append(B[j])
            j += 1

    while i < len(A):
        iteracoes += 1
        resultado.append(A[i])
        i += 1

    while j < len(B):
        iteracoes += 1
        resultado.append(B[j])
        j += 1

    return resultado

def merge_sort(L):
    global iteracoes
    iteracoes += 1

    if len(L) == 1:
        return L

    meio = len(L) // 2
    A = L[:meio]
    B = L[meio:]

    A = merge_sort(A)
    B = merge_sort(B)

    return merge(A, B)

def testar(tamanho):
    global iteracoes
    vetor = [random.randint(-10**9, 10**9) for _ in range(tamanho)]

    iteracoes = 0
    inicio = time.perf_counter()
    ordenado = merge_sort(vetor)
    fim = time.perf_counter()

    print(f"Tamanho do vetor: {tamanho}")
    print(f"Primeiros 10 elementos ordenados: {ordenado[:10]}")
    print(f"Iterações: {iteracoes}")
    print(f"Tempo gasto: {fim - inicio:.6f} segundos")
    print("-" * 40)

# testes pedidos
testar(32)
testar(2048)
testar(1048576)
```
