```python
import random
import time

iteracoes = 0

def maxVal2(A, init, end):
    global iteracoes
    iteracoes += 1

    if end - init <= 1:
        return max(A[init], A[end])
    else:
        m = (init + end) // 2
        v1 = maxVal2(A, init, m)
        v2 = maxVal2(A, m + 1, end)
        return max(v1, v2)

def testar(tamanho):
    global iteracoes
    A = [random.randint(-10**9, 10**9) for _ in range(tamanho)]

    iteracoes = 0
    inicio = time.perf_counter()
    maior = maxVal2(A, 0, len(A) - 1)
    fim = time.perf_counter()

    print(f"Tamanho do vetor: {tamanho}")
    print(f"Maior valor: {maior}")
    print(f"Iterações/chamadas: {iteracoes}")
    print(f"Tempo gasto: {fim - inicio:.6f} segundos")
    print("-" * 40)

testar(32)
testar(2048)
testar(1048576)
```
